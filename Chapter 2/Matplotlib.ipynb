```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.arange(10)
print(data)

plt.plot(data)
plt.title("Sample line plot")
```

### Customizing figure 

```python
plt.figure(figsize=(6, 6), dpi=100, facecolor='lightgray')
plt.plot(data)
```

### Adding subplots in a figure

### single subplot in a figure

```python
fig = plt.figure()  # Create a new figure
ax1 = fig.add_subplot(1, 1, 1)  # 1 row, 1 column, 1st subplot
ax1.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

### Multiple subplots

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
```

```python
fig = plt.figure(figsize=(8, 6))

# Add 1st subplot (1st row, 2nd column, index=1)
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot([1, 2, 3], [4, 5, 6], color='red')
ax1.set_title("Plot 1")

# Add 2nd subplot (1st row, 2nd column, index=2)
ax2 = fig.add_subplot(2, 2, 2)
ax2.bar([1, 2, 3], [6, 5, 4], color='blue')
ax2.set_title("Plot 2")

# Add 3rd subplot (2nd row, full width, index=3)
ax3 = fig.add_subplot(2, 1, 2)
ax3.scatter([1, 2, 3], [4, 5, 6], color='green')
ax3.set_title("Plot 3")

#plt.tight_layout()  # Adjusts spacing between plots
#plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)
plt.show()
```

## Colors, Markers, and Line Styles

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(np.random.standard_normal(30).cumsum(), color="green", linestyle="dashed", marker="o")
plt.show()
```

## Ticks, Labels, and Legends

### Setting the title, axis labels, ticks, and tick labels

```python
fig, ax = plt.subplots()
ax.plot(np.random.standard_normal(1000).cumsum())

#To change the x-axis ticks, use 
  #set_xticks() - instructs matplotlib where to place the ticks along the data range, by default these locations will also be the labels
  #set_xticklabels() - used to set any other values as the labels
ticks = ax.set_xticks([0, 250, 500, 750, 1000])

labels = ax.set_xticklabels(["one", "two", "three", "four", "five"], rotation=30, fontsize=8)

ax.set_title("My first matplotlib plot")  # Setting plot title
```

### Adding legends

```python
fig, ax = plt.subplots()

ax.plot(np.random.randn(1000).cumsum(), color="black", label="one");
ax.plot(np.random.randn(1000).cumsum(), color="black", linestyle="dashed", label="two");
ax.plot(np.random.randn(1000).cumsum(), color="black", linestyle="dotted", label="three");

ax.legend()
```

### Annotations and Drawing on a Subplot

```python
# Create a simple plot
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])

# Add text to the plot
plt.text(2, 22, "This is a point", fontsize=12, color='red')

plt.show()
```

```python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])

# Add an arrow pointing to (2, 20)
plt.arrow(2, 20, 0.5, 3, head_width=0.1, head_length=0.5, color='blue')

plt.show()
```

```python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])

# Annotate point (2, 20)
plt.annotate("Important Point", xy=(2, 20), xytext=(1.5, 25),
             arrowprops=dict(arrowstyle="->", color='green'))

plt.show()
```
