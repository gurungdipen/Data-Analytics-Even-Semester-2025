# Data Visualization with Pandas - Part 1

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
plt.style.use("fivethirtyeight")
```

## Plotting graph with plot() method

```python
data = pd.Series([100, 170, 200, 210, 300], index=["Mon", "Tue", "Wed", "Thu", "Fri"])
print(data)
```

```python
# Plot the Series
data.plot(title="Sales Over the Week")
plt.ylabel("Sales ($)")
plt.show()
```

```python
# Sample Data
df = pd.DataFrame({
    'Year': [2018, 2019, 2020, 2021, 2022],
    'Sales_A': [100, 200, 250, 400, 450],
    'Sales_B': [150, 180, 210, 320, 500]
})
print(df)
```

```python
# Line Plot
df.plot(x='Year', y=['Sales_A', 'Sales_B'], kind='line', marker='o', title="Yearly Sales")  
plt.show()
```

## Bar Charts

```python
df.plot(x='Year', y='Sales_A', kind='bar')
plt.show()
```

```python
df.plot(x='Year', y=['Sales_A', 'Sales_B' ], kind='bar', stacked=True)
plt.show()
```

## Histograms

```python
iris=pd.read_csv("iris.csv", header=None)
iris.columns=["sepal_length","sepal_width", "petal_length", 
              "petal_width", "species"]
```

You can access this data set [here](https://github.com/tirendazacademy/data-visualization-with-python).

```python
iris.head()
```

```python
iris.plot.hist(alpha=0.7)
plt.show()
```

```python
iris.plot.hist(alpha=1, stacked=True)
plt.show()
```

```python
bins=25
```

```python
iris.plot.hist(alpha=1, stacked=True, bins=25)
plt.show()
```

```python
iris["sepal_width"].plot.hist(orientation="horizontal")
#iris["sepal_width"].plot(kind='hist', orientation="horizontal")
plt.show()
```

```python
iris.hist(color="blue", alpha=1, bins=20)
plt.show()
```

```python
iris.hist("petal_length",by="species")
plt.show()
```

## Boxplot charts

```python
iris.plot.box() 
plt.show()
```

```python
colors={'boxes': 'Red', 'whiskers': 'blue','medians': 'Black', 'caps': 'Green'}
```

```python
iris.plot.box(color=colors)
```

```python
iris.plot.box(vert=False)
```

```python
iris.boxplot()
```

```python
plt.rcParams["figure.figsize"]=(8,8)
plt.style.use("ggplot")
iris.boxplot(by='species')
```

```python
# Sample Data with an Outlier
df = pd.DataFrame({'Scores': [50, 55, 60, 65, 100]})

# Box Plot
df.boxplot(column=['Scores'])
plt.title("Box Plot Identifying Outliers")
plt.show()
```
