### Seaborn is a powerful Python library for statistical data visualization built on top of Matplotlib. 
### Seaborn provides data visualizations that are typically more aesthetic and statistically sophisticated.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset
df = sns.load_dataset("tips")

# Display the first 5 rows
print(df.head())
```

#### Scatter plot

```python
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.show()
```

```python
sns.barplot(x="tip", y="day", data=df, orient="h", hue="time")
```

## NOTE: Plotting functions in seaborn take a data argument, which can be a pandas DataFrame. x and y refer to column names. The other arguments are used for customization.

#### Lineplot

```python
sns.lineplot(x="size", y="total_bill", data=df)
plt.show()
```

#### Histogram

```python
sns.histplot(df["total_bill"], bins=20, kde=True)
plt.show()
```

#### Boxplot

```python
sns.boxplot(x="day", y="total_bill", data=df)
plt.show()
```

#### Countplot

```python
sns.countplot(x="day", hue="sex", data=df)
plt.show()
```

#### Heatmap

```python
import numpy as np

# Compute the correlation matrix
corr = df.corr(numeric_only=True)

# Create a heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.show()
```

### NOTE: Red indicates strong correlation, blue indicates a comparatively weaker positive correlation.

```python
sns.pairplot(df, hue="sex")
plt.show()
```
